```python

# -*- coding: utf-8 -*-
"""
   - :   
"""
import numpy as np
import matplotlib.pyplot as plt

# 1.   
ell = np.arange(2, 2501)  #   
Dl_standard = 1000 * ell * (ell + 1) / (2 * np.pi)  #  CMB 
error = Dl_standard * 0.05  #  

# 2.    (   )
def quantum_correction(ell, Dl, beta=9.2e3, gamma=5.2e-8):
    return Dl * (1 + beta*(ell/1000)**2*np.exp(-ell/2000) + gamma*np.log(ell+1))

Dl_quantum = quantum_correction(ell, Dl_standard)

# 3.   
plt.figure(figsize=(12,7), dpi=120)
plt.errorbar(ell[::50], Dl_standard[::50], yerr=error[::50],
             fmt='o', label=' ', alpha=0.7)
plt.plot(ell, Dl_quantum, 'r-', linewidth=2,
         label=' -\n:   ')
plt.xscale('log')
plt.yscale('log')
plt.xlabel(' (  )', fontsize=14)
plt.ylabel('D [K]', fontsize=14)
plt.title('     ', pad=20)
plt.legend(loc='lower left')
plt.grid(True, which="both", ls="--")
plt.savefig('CMB_Comparison_Ahmed_Ait_Zaouit.png', bbox_inches='tight')
plt.show()

print("       CMB_Comparison_Ahmed_Ait_Zaouit.png")
```
